```python
import numpy as np

def remove_outliers(numbers, column):
    threshold = 0.4
    foo = np.array(numbers)[:,column]
    mean = np.mean(foo)
    std = np.std(foo)
    z_scores = np.abs(foo - mean) < threshold * std
    return numbers[z_scores]
```

```python
import pandas as pd
df = pd.read_csv('dataset.csv')
df = remove_outliers(df, 2)
import plotly.express as px
fig = px.density_mapbox(df, lat='latitude', lon='longitude', z='preco', radius=50,
                        center=dict(lat=-20.36, lon=-40.27), zoom=11,
                        mapbox_style="stamen-terrain")
fig.show()
```
